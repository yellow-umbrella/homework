```python
import numpy as np
from sympy import Matrix
industry = [800, 400, 800]
agro = [300, 350, 600]
y = Matrix([[1100, 850]])
matrix = Matrix([industry, agro])
matrix
```

```python
norm = np.sum(matrix, axis=1)
print("Normal:")
Matrix(norm)
```

```python
A = Matrix(matrix[:, 0:2]/norm)
A.n(2)
```

 : X = AX + y  X = y(E  A)<sup>1</sup>

```python
X = y * (Matrix.eye(2) - A).inv()
print("   :", X[0].n(6), ". ")
print("   :", X[1].n(6), ". ")
```

```python
B = np.linalg.inv(np.eye(3) - A)
Matrix(B).n(3)
```

```python
seq = np.zeros((n, n));
k = -1
eps = np.array([[0.01 for _ in range(n)] for _ in range(n)])
A_prev = np.eye(n)
while np.any(abs(np.subtract(B, seq)) >= eps):
 seq = np.add(seq, A_prev)
 A_prev = np.dot(A_prev, A)
 k += 1
print(f"  ({k} ):")
Matrix(seq).n(4)
```
