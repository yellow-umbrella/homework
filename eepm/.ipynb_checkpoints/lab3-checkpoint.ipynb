```python
import numpy as np
from sympy import Matrix
industry = [800, 400, 800]
agro = [300, 350, 600]
y = Matrix([[1100, 850]])
matrix = Matrix([industry, agro])
matrix
```

```python
norm = np.sum(matrix, axis=1)
print("Normal:")
Matrix(norm)
```

```python
A = Matrix(matrix[:, 0:2]/norm)
A.n(2)
```

Повний випуск: X = AX + y ⇒ X = y(E − A)<sup>−1</sup>

```python
X = y * (Matrix.eye(2) - A).inv()
print("Повний випуск промислової галузі:", X[0].n(6), "млн. грн")
print("Повний випуск сільськогосподарської продукції:", X[1].n(6), "млн. грн")
```

```python
A = np.array([
 [0.4, 0.1, 0.5],
 [0.1, 0.5, 0.3],
 [0.5, 0.3, 0.2] 
])
n = 3
eig, vt = np.linalg.eig(A.T)
eig, v = np.linalg.eig(A)
print("Власні числа:")
Matrix(eig).n(4)
```

```python
poly = Matrix(A).charpoly().as_expr().n(3)
print("Характеристичний поліном:")
poly
```

```python
f = max(abs(val) for val in eig)
rv = v[:, np.where(eig == f)[0]]
lv = vt[:, np.where(eig == f)[0]]
print("Число Фробеніуса:", f)
```

```python
print("Правий вектор:")
Matrix(rv).n(3)
```

```python
print("Лівий вектор:")
Matrix(lv).n(3)
```

```python
if (abs(f) < 1):
 print("Висновок: продуктивна")
else:
 print("Висновок: непродуктивна")
```

Матриця повних витрат: B = (E − A)<sup>−1</sup>

```python
B = np.linalg.inv(np.eye(3) - A)
Matrix(B).n(3)
```

```python
seq = np.zeros((n, n));
k = -1
eps = np.array([[0.01 for _ in range(n)] for _ in range(n)])
A_prev = np.eye(n)
while np.any(abs(np.subtract(B, seq)) >= eps):
 seq = np.add(seq, A_prev)
 A_prev = np.dot(A_prev, A)
 k += 1
print(f"Ряд сходиться ({k} ітерацій):")
Matrix(seq).n(4)
```

```python
s = [0.4, 0.3, 0.6]
p = s @ np.linalg.inv(np.eye(3) - A)
print("Вектор цін:")
Matrix(p).n(4)
```
